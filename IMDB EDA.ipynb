### Top 20 performance of different genres based on Average Rating criteria.

```python
# Set the number of top genres you want to analyze.
top_n_genres = 20  # Change this value to the desired number of top genres.

# Group the data by genre and calculate the average ratings for each genre.
genre_stats = data.groupby('Genre')['Rating'].mean().reset_index()

# Sort the genres by average rating in descending order and select the top N.
top_genres = genre_stats.sort_values(by='Rating', ascending=False).head(top_n_genres)

# Set up the figure and axis for plotting.
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top_genres['Genre'], y=top_genres['Rating'], palette="Blues")
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)  # Rotate x-axis labels for readability.
plt.title(f"Top {top_n_genres} Genres by Average Rating")
plt.xlabel("Genre")
plt.ylabel("Average Rating")

plt.tight_layout()
plt.show()
```

### Top 20 Movie Genres by Frequency

```python
# Count the frequency of each genre.
genre_counts = data['Genre'].value_counts()

# Select the top 20 most frequent genres to visualize.
top_n_genres = 20  # Change this value to the desired number of top genres.
top_genres = genre_counts.head(top_n_genres)

# Plot a bar chart to visualize the frequency of the top 20 genres.
plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.index, y=top_genres.values, palette="Blues")
plt.title(f"Top {top_n_genres} Movie Genres by Frequency")
plt.xlabel("Genre")
plt.ylabel("Frequency")
plt.xticks(rotation=90)  # Rotate x-axis labels for readability

plt.tight_layout()
plt.show()
```

### Classifying Movies Based on Ratings [Excellent, Good and Average]

```python
def rating(rating):
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

### Finding Unique Values From Genre 

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
unique_list=[]
for item in list1:
    for item1 in item:
        if item1 not in unique_list:
            unique_list.append(item1) 
```
